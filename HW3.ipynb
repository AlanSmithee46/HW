```python
import cv2 as cv
from skimage import data
import numpy as np
import matplotlib.pyplot as plt

def hw3q1(image, n):
    rectangle = []
    h = image.shape[0]
    w = image.shape[1]
    for i in range(n):
        r1 = np.random.randint(0, h)
        c1 = np.random.randint(0, w)
        r2 = np.random.randint(r1, h)
        c2 = np.random.randint(c1, h)
        rectangle.append([r1,c1,r2,c2])
    return np.array(rectangle)
```

```python
image = data.astronaut()
image[0]
```

```python
hw3q1(image, 4)
```

#### Question 2

```python
def red_square(image,r1,c1,r2,c2):
    for i in range(r1,r2):
        for j in range(c1,c2):
            image[i,j,0] = 255
    return image
```

```python
plt.imshow(red_square(data.astronaut(), 10, 10, 400, 400))
```
